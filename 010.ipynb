```python
# Part 1
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# Part 2
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [5, 7, 6, 8, 10]

plt.scatter(x, y, s=100, color="green")
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# Part 3
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]

plt.plot(x, y1, label="Linear", marker="o")
plt.plot(x, y2, label="Quadratic", linestyle="--", marker="s")

plt.title("Multiple Lines")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Part 4
import matplotlib.pyplot as plt

categories = ["A", "B", "C"]
values = [10, 20, 15]

plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()

langs = ["Python", "Java", "C++"]
shares = [40, 35, 25]

plt.pie(shares, labels=langs, autopct="%1.1f%%")
plt.title("Pie Chart Example")
plt.show()
```

```python
# Part 5
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 6)
y1 = x
y2 = x**2

plt.figure(figsize=(6, 6))

plt.subplot(2, 1, 1)
plt.plot(x, y1, marker="o")
plt.title("First Subplot")
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(x, y2, marker="s")
plt.title("Second Subplot")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Part 6
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=20)
plt.title("Histogram Example")
plt.savefig("histogram_example.png")
plt.show()

plt.boxplot(data)
plt.title("Box Plot Example")
plt.show()
```
